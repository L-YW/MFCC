```python
from scipy import signal
import scipy.fftpack as fft
import matplotlib.pyplot as plt
```

```python
window=signal.boxcar(51)
plt.plot(window)
plt.title("Boxcar window")
plt.ylabel("Amplitude")
plt.xlabel("Sample")
```
